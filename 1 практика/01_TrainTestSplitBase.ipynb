#   1.Разбиение на тестовую и обучающую выборку в Python

Цель:

Провести анализ сбалансированности классов в наборе данных "spambase.csv" и разбить его на обучающую и тестовую выборки в соотношениях 80/20 и 70/30, сохраняя стратификацию.

Данные:

Набор данных "spambase.csv", содержащий 4601 экземпляр писем, разделенных на два класса: "нормальная почта" (0) и "спам" (1).

```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit

# Загрузка данных из файла csv
data = pd.read_csv('spambase.csv')

# 1. Создание pandas dataframe на основе "сырых" данных
df = pd.DataFrame(data)

# 2. Анализ сбалансированности классов
class_counts = df['class'].value_counts()
print(class_counts)

# 3. Разбиение на тестовую обучающую выборку
sss_80_20 = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
sss_70_30 = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)

for train_index, test_index in sss_80_20.split(df, df['class']):
    train_set_80_20 = df.loc[train_index]
    test_set_80_20 = df.loc[test_index]

for train_index, test_index in sss_70_30.split(df, df['class']):
    train_set_70_30 = df.loc[train_index]
    test_set_70_30 = df.loc[test_index]
```

Наблюдается дисбаланс классов: "нормальная почта" составляет 60,6% от всего набора данных, а "спам" - 39,4%.

Визуализация:

```python
import matplotlib.pyplot as plt

# Визуализация сбалансированности классов
class_counts.plot(kind='bar')
plt.title('Сбалансированность классов')
plt.xlabel('Класс')
plt.ylabel('Количество')
plt.show()

# Визуализация разбиения на обучающую и тестовую выборки
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

train_set_80_20['class'].value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title('Обучающая выборка 80/20')

train_set_70_30['class'].value_counts().plot(kind='bar', ax=axes[1])
axes[1].set_title('Обучающая выборка 70/30')

plt.show()
```

Вывод:

В наборе данных "spambase.csv" наблюдается дисбаланс классов.
Были созданы обучающая и тестовая выборки в соотношениях 80/20 и 70/30, сохраняя стратификацию.

Дальнейшие шаги:

Можно использовать методы балансировки классов, например, передискретизацию или понижение дискретизации, а также применить алгоритмы машинного обучения для классификации писем как "нормальная почта" или "спам".
