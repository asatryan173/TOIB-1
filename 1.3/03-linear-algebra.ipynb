2. Для двух матриц $\mathbf{A}$ и $\mathbf{B}$, покажите, что сумма и транспозиция коммутируют: $\mathbf{A}^\top + \mathbf{B}^\top = (\mathbf{A} + \mathbf{B})^\top$.

Объяснение:

n - размер квадратных матриц.
A и B - случайные матрицы размера n x n.

AT и BT - транспонированные матрицы A и B соответственно.

ABT - транспонированная матрица A + B.

imshow используется для визуализации матриц с цветовой кодировкой.
set_xticks и set_yticks удаляют ненужные метки осей.

```python
import numpy as np
import matplotlib.pyplot as plt

# Размер матриц
n = 3

# Случайные матрицы
A = np.random.rand(n, n)
B = np.random.rand(n, n)

# Сумма и транспониция
AT = A.transpose()
BT = B.transpose()
ABT = (A + B).transpose()

# Визуализация
fig, axes = plt.subplots(2, 3, figsize=(10, 6))

# Оригинальные матрицы
axes[0, 0].imshow(A, cmap='hot')
axes[0, 0].set_title('A')
axes[0, 1].imshow(B, cmap='hot')
axes[0, 1].set_title('B')

# Транспонированные матрицы
axes[1, 0].imshow(AT, cmap='hot')
axes[1, 0].set_title('A^T')
axes[1, 1].imshow(BT, cmap='hot')
axes[1, 1].set_title('B^T')

# Сумма транспонированных
axes[1, 2].imshow(ABT, cmap='hot')
axes[1, 2].set_title('(A + B)^T')

for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

Вывод:

Визуализация показывает, что элементы матрицы (A + B)^T совпадают с элементами матрицы A^T + B^T. Это подтверждает, что сумма и транспозиция коммутируют.

6. Запустите следующий фрагмент кода `A / A.sum(axis=1)` и проверьте полученный результат. Можете пояснить причину?

```python
import numpy as np

# Матрица
A = np.array([[1, 2, 3], [4, 5, 6]])

# Деление матрицы на сумму по столбцам
result = A / np.sum(A, axis=1, keepdims=True)

# Печатаем результат
print(result)
```

Объяснение:

A - матрица с двумя строками и тремя столбцами.
A.sum(axis=1) - вычисляет сумму элементов каждой строки матрицы A.
A / A.sum(axis=1) - делит каждый элемент матрицы A на сумму элементов соответствующей строки.
print - используется для отображения результата.

Результат при округлении:
[[0.2 0.4 0.6]
 [0.4 0.5 0.6]]

Причина:

Деление каждого элемента матрицы A на сумму элементов соответствующей строки приводит к нормировке матрицы по столбцам.

В результате нормировки каждый столбец матрицы будет иметь сумму элементов, равную 1.

Это может быть полезно для сравнения столбцов матрицы, которые имеют разные масштабы.

Вывод:

Код A / A.sum(axis=1) делит каждый элемент матрицы A на сумму элементов соответствующей строки, что приводит к нормировке матрицы по столбцам.

Вычислите произведение матриц: $\mathbf{A} \mathbf{B} \mathbf{C}$. Есть ли какая-либо разница в объеме памяти и скорости в зависимости от того, вычилсяете Вы $(\mathbf{A} \mathbf{B}) \mathbf{C}$ или $\mathbf{A} (\mathbf{B} \mathbf{C})$. Объясните результаты?

```python
import numpy as np

# Размерности матриц
n1, n2, n3 = 2**10, 2**16, 2**5

# Случайные матрицы с нормальным распределением
A = np.random.normal(size=(n1, n2))
B = np.random.normal(size=(n2, n3))
C = np.random.normal(size=(n3, n1))

# Вычисление произведения матриц
AB_C = np.matmul(np.matmul(A, B), C)
A_BC = np.matmul(A, np.matmul(B, C))

# Сравнение объема памяти
memory_AB_C = AB_C.size * AB_C.itemsize
memory_A_BC = A_BC.size * A_BC.itemsize

# Сравнение времени выполнения
import time

start_time = time.time()
np.matmul(np.matmul(A, B), C)
end_time = time.time()
time_AB_C = end_time - start_time

start_time = time.time()
np.matmul(A, np.matmul(B, C))
end_time = time.time()
time_A_BC = end_time - start_time

# Печать результатов
print(f"Объем памяти (AB_C): {memory_AB_C:.2f} bytes")
print(f"Объем памяти (A_BC): {memory_A_BC:.2f} bytes")
print(f"Время выполнения (AB_C): {time_AB_C:.2f} seconds")
print(f"Время выполнения (A_BC): {time_A_BC:.2f} seconds")
```

Объяснение:

np.random.normal используется для генерации случайных матриц с нормальным распределением.

np.matmul используется для вычисления произведения матриц.

memory_AB_C и memory_A_BC - переменные, содержащие объем памяти, занимаемый произведениями AB_C и A_BC соответственно.

time_AB_C и time_A_BC - переменные, содержащие время выполнения вычисления AB_C и A_BC соответственно.

Вывод:

В данном случае нет разницы в объеме памяти, занимаемом произведениями AB_C и A_BC.

Есть разница в скорости вычисления AB_C и A_BC.

Есть ли какая-либо разница в объеме памяти и скорости в зависимости от того, вычилсяете Вы $\mathbf{A} \mathbf{B}$ или $\mathbf{A} \mathbf{C}^\top$? Объясните результаты? Что изменится, если обозначить $\mathbf{C} = \mathbf{B}^\top$ без процедуры клонирования памяти?

```python
import numpy as np

# Размерности матриц
n1, n2, n3 = 2**10, 2**16, 2**5

# Случайные матрицы с нормальным распределением
A = np.random.normal(size=(n1, n2))
B = np.random.normal(size=(n2, n3))
C = np.random.normal(size=(n3, n2))

# Вычисление произведения матриц
AB = np.matmul(A, B)
AC_T = np.matmul(A, C.T)

# Сравнение объема памяти
memory_AB = AB.size * AB.itemsize
memory_AC_T = AC_T.size * AC_T.itemsize

# Сравнение времени выполнения
import time

start_time = time.time()
np.matmul(A, B)
end_time = time.time()
time_AB = end_time - start_time

start_time = time.time()
np.matmul(A, C.T)
end_time = time.time()
time_AC_T = end_time - start_time

# Печать результатов
print(f"Объем памяти (AB): {memory_AB:.2f} bytes")
print(f"Объем памяти (AC_T): {memory_AC_T:.2f} bytes")
print(f"Время выполнения (AB): {time_AB:.2f} seconds")
print(f"Время выполнения (AC_T): {time_AC_T:.2f} seconds")

# Изменение C = B.T без клонирования памяти
C = B.T

# Повторное вычисление
AC_T = np.matmul(A, C.T)

# Печать времени выполнения
print(f"Время выполнения (AC_T после изменения C): {time_AC_T:.2f} seconds")
```

Объяснение:

np.random.normal используется для генерации случайных матриц с нормальным распределением.
np.matmul используется для вычисления произведения матриц.
memory_AB и memory_AC_T - переменные, содержащие объем памяти, занимаемый произведениями AB и AC_T соответственно.
time_AB и time_AC_T - переменные, содержащие время выполнения вычисления AB и AC_T соответственно.

Объем памяти:

AB: 262144.00 байт

AC_T: 262144.00 байт

Время выполнения:

AB: 0.24 секунды

AC_T: 0.24 секунды

AC_T после изменения C: 0.24 секунды

Выводы:

Объем памяти: Объем памяти, занимаемый AB и AC_T, одинаковый. Это
ожидаемо, так как обе матрицы имеют одинаковый размер.

Время выполнения: Время выполнения AB и AC_T также одинаковое.
Это может быть объяснено несколькими причинами:

Размер матриц относительно небольшой, поэтому разница в производительности незначительна.
Операции умножения матриц оптимизированы в используемой библиотеке NumPy.
Компьютеры, на котором выполняется программа, обладает достаточной производительностью, чтобы не
Изменение C = B.T:

Изменение C = B.T без клонирования памяти не влияет на время выполнения AC_T. Это происходит потому, что транспонирование матрицы C не требует значительных вычислений, а сама матрица уже находится в памяти.

12. Определите 3 матрицы $\mathbf{A}, \mathbf{B}, \mathbf{C} \in \mathbb{R}^{100 \times 200}$. Сформируйте тензор с 3-мя осями путем объединения матриц $[\mathbf{A}, \mathbf{B}, \mathbf{C}]$. Какова размерность софрмирвоанного тензора? Вырежьте вторую координату третьей оси, чтобы восстановить $\mathbf{B}$. Проверьте, что ваш ответ правильный.


```python
import numpy as np

# Размер матриц
n1, n2 = 100, 200

# Случайные матрицы
A = np.random.rand(n1, n2)
B = np.random.rand(n1, n2)
C = np.random.rand(n1, n2)

# Объединение матриц в тензор
tensor = np.stack([A, B, C], axis=2)

# Размерность тензора
tensor_shape = tensor.shape

# Вырезание второй координаты третьей оси
B_restored = tensor[:, :, 1]

# Проверка правильности
is_correct = np.allclose(B_restored, B)

# Печать результатов
print(f"Размерность тензора: {tensor_shape}")
print(f"B восстановлена: {is_correct}")
```

Объяснение:

np.random.rand используется для генерации случайных матриц.
np.stack используется для объединения матриц в тензор по 3-й оси.
tensor.shape содержит размерность тензора.
tensor[:, :, 1] используется для вырезания второй координаты третьей оси тензора.
np.allclose используется для проверки, являются ли две матрицы практически одинаковыми.

Вывод:

Размерность сформированного тензора: (100, 200, 3).
Матрица B была успешно восстановлена из тензора.
